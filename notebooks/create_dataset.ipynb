```python
def pdk_scatter_layer(gdf, color):
    return pdk.Layer(
        "ScatterplotLayer",
        data=gdf[["name", "coords"]].copy(),
        stroked=True,
        filled=True,
        pickable=True,
        auto_highlight=True,
        get_position="coords",
        get_fill_color=color,
        get_line_color=[0, 0, 0],
        get_radius=500,
        radius_min_pixels=3,
        radius_max_pixels=20,
    )
```

```python
import email
from email import policy
from email.parser import BytesParser
from pathlib import Path
import openai
import numpy as np
import pandas as pd
from jinja2 import Template


def get_email_text(email_path):
    with open(email_path, 'rb') as f:
        msg = BytesParser(policy=policy.default).parse(f)
    return msg.get_body(preferencelist=('plain')).get_content()

emails = [get_email_text(fp) for fp in Path('./data/emails').glob('**/*.eml')]
emails = [email for email in emails if len(email) < 10_000 and len(email) > 100]
```

```python
SYSTEM_MSG = dict(role='system', content=Template("""You are a generating sample data for testing a system that automatically processes emails from clients with shipping requests of cargo.

The requests should include port of lading (POL), port of discharge (POD), Pickup date range and the customer name.

Use some of the following ports as POL and POD: {{ ports}}

Return output on the following form:

{
metadata: {
"POL":  "",
"POD":  "",
"date_range":  {"start_date": "", "end_date": ""},
"customer": ""
},
"email_request": ""
}""").render(ports=str(pd.Series(ports).sample(20).tolist())))

shipping_data = []
for i in range(30):
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[SYSTEM_MSG],
        temperature=1,
    )
    shipping_requests = response.choices[0].message.content.split("\n\n")
    shipping_data.extend(shipping_requests)
```
